# **5. Merging Credit_card table and Credit_card_label table uding Ind_id column**

```python
!pip install pandas
```

```python
label_data=pd.read_csv('Credit_card_label.csv')
```

```python
label_data[label_data.duplicated()]
```

```python
label_data.columns=[col.capitalize() for col in label_data.columns]
```

```python
label_data.isnull().sum()
```

```python
label_data
```

```python
merged_data = pd.merge(df, label_data, on='Ind_id')
```

```python
merged_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Separate the features Variable (Ind_id) and target Variable (Label)
x = merged_data[['Ind_id']]
y = merged_data['Label']
```

```python
# Convert the categorical feature to numerical using one-hot encoding
#x = pd.get_dummies(x, columns=['Car_owner'], drop_first=True)
```

```python
# Split the data into training anf testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# **6. Modelling**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
# Make Preditions on the testing data
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Specify zero_division parameter
zero_division_option = "warn"

precision = precision_score(y_test, y_pred, zero_division=zero_division_option)
recall = recall_score(y_test, y_pred, zero_division=zero_division_option)
f1 = f1_score(y_test, y_pred, zero_division=zero_division_option)
```

```python
# Evaluate the model Performance
from sklearn.metrics import accuracy_score
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
print('Precision: ', precision)
print('Recall: ', recall)
print('F1 Score: ', f1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
